# Task
Identify an animal using an expert system based on the provided initial facts and rules.

## Define Rule Application Function

### Subtask:
Define a Python function, `apply_rules`, that takes the current set of facts and a list of rules. This function will iterate through the rules, check if their 'if' conditions are met by the current facts, and if so, add the 'then' facts to the current set. It should continue applying rules until no new facts are derived in a full pass.


**Reasoning**:
The subtask requires defining a Python function `apply_rules` according to the provided specifications. This function will iteratively apply rules to a set of facts until no new facts are derived.



```python
def apply_rules(current_facts, rules):
    while True:
        new_fact_added = False
        for rule in rules:
            conditions_met = True
            # Check 'if' conditions
            for fact, value in rule['if'].items():
                if fact not in current_facts or current_facts[fact] != value:
                    conditions_met = False
                    break

            # If 'if' conditions are met, apply 'then' conditions
            if conditions_met:
                for fact, value in rule['then'].items():
                    if fact not in current_facts or current_facts[fact] != value:
                        current_facts[fact] = value
                        new_fact_added = True

        # If no new facts were added in this pass, break the loop
        if not new_fact_added:
            break
    return current_facts

print("The 'apply_rules' function has been defined.")
```

## Run Inference Engine

### Subtask:
Execute the `apply_rules` function using the `initial_facts` and the `rules` list available in the kernel. This will derive new facts based on the defined knowledge base.


**Reasoning**:
The subtask requires executing the `apply_rules` function with `initial_facts` and `rules`, storing the result in `derived_facts`, and then printing it.
